```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
R = 5
a = np.pi/3
x = R * np.cos(a)
y = R * np.sin(a)
print(f'Декартовы координаты: ({x}, {y})')
```

```python
a = np.linspace(0, 2 * np.pi, 1000)
R = 5
y = 2*R*np.cos(a)
plt.polar(a, y)
plt.show()
```

```python
a = np.linspace(0, 2 * np.pi, 10)
P = 5
b = np.pi/6
y = P/np.cos(a - b)
plt.polar(a, y)
plt.show()
```
