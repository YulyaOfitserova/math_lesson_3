```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2, 5, 201)
plt.plot(x, (np.exp(x) - 1 + x)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 7)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(equations, (-1, 0))
print(x1, y1)
x2, y2 = fsolve(equations, (2, 3))
print(x2, y2)
```

```python
#решение системы с неравенством на рисунке
x_ = np.linspace(x1, 0, 201) 
plt.plot(x_, x_**2 - 1)
x0 = np.linspace(x2, 5, 201) 
plt.plot(x0, x0**2 - 1)
```
